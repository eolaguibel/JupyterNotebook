# Muestra los países con esperanza de vida femenina mayor al promedio general y sus datos 

```python
#Muestra los países con esperanza de vida femenina mayor al promedio general y sus datos
#Indicator 'Life expectancy'

#indexar en base al indicador 'Indicator'
IndexEsperanzaVidaFemenina=EsperanzaVida.set_index(['Inequality'])
```

```python
#Muestro el dataset para ver los cambios
IndexEsperanzaVidaFemenina
```

```python
#Muestro los datos estadisticos de la esperanza de vida femenina en 39 paises
EsperanzaVidaFemenina.describe()
```

```python
promvidamujeres=round(sum(EsperanzaVidaFemenina['Value'])/len(EsperanzaVidaFemenina),2)
print("El promedio de vida de mujeres es {} años".format(promvidamujeres))
```

```python
#Muestra los países con esperanza de vida femenina mayor al promedio general y sus datos

#comparo todos los registros de la columna 'value' del dataset EsperanzaVidaFemenina para que solo 
#filtre los que son mayores al promedio, que en este caso es 82.24
Esp_Vida_arriba_Prom = EsperanzaVidaFemenina[EsperanzaVidaFemenina['Value'] > 82.24 ]

#presento la información en pantalla, bonita y explicada
print("Existen {} paises cuya esperanza de vida para mujeres supera el promedio".format(len(Esp_Vida_arriba_Prom)))
print("A continuación de muestran ordenados de mayor a menor")

#ordeno el dataset con sort_values, checar la sintaxis del sort
Esp_Vida_arriba_Prom.sort_values(by=['Value'],axis=0,ascending=False)
```

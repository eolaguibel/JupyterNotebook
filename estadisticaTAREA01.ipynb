```python
# ejercicio 2.3
## Código Ejercicio 2.3 
%reset
#importar libreria numpy
import numpy as np
#inicializo vriables
pasos=0
k=2000 #veces_experim
n=1002 #tiros_experim
lista=[]
dif_pos_ini=[]
#genero numero aleatorio
for veces_experim in range(int(k)):
    for tiros_experim in range(int(n)):
        moneda=np.random.rand()
        if moneda>0.5:
            pasos+=1
        else:
            pasos-=1
        if tiros_experim<15:
            lista.append(pasos)
    print("posicion inicial=0,0 posicion final={} luego de 1002 pasos aleatorios".format(pasos))
    dif_pos_ini.append(pasos)
    print(np.array(lista))
    pasos=0
    lista.clear()
```

```python
    
## Código Ejercicio 2.4 
%reset
import numpy as np
#inicializo vriables
pasos=0
k=2000 #veces_experim
n=1002 #tiros_experim
lista=[]
dif_pos_ini=[]
#genero numero aleatorio
for veces_experim in range(int(k)):
    for tiros_experim in range(int(n)):
        moneda=np.random.rand()
        if moneda>0.5:
            pasos+=1
        else:
            pasos-=1
        if tiros_experim<15:
            lista.append(pasos)
    print("posicion inicial=0,0 posicion final={} luego de 1002 pasos aleatorios".format(pasos))
    dif_pos_ini.append(pasos)
    a= np.array(lista)
    #imprimo mi lista de 15 n
    print(a)
    #asignar a las variables recien creadas su valor usano funciones de numpy
    media= np.mean(a)
    mediana=np.median(a)
    desv=np.std(a)
    #moda=stats(a)
    print("media:{}, mediana:{}, desviación estandar:{}, rango:{} y {}".format(media,mediana,desv,min(a),max(a)))
    pasos=0
    lista.clear()
```
